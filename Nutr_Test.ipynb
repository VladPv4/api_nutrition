# Function - API > Excel

```python
def query_excel(query):
    base_url = "https://trackapi.nutritionix.com//v2/search/instant"
    headers = {
        'x-app-id': '43fe6470',
        'x-app-key': 'bbb256572f595d38693c0c5dc462d5be'}
    body = {
        'query': query}
    r2 = requests.post(base_url, params=headers, data=body)
    r2_json = r2.json()
    df = pd.DataFrame()
    for i in r2_json:
        for l in r2_json[i]:
            df = df.append(l, ignore_index=True)
    return df
#     df.to_excel('output.xlsx')
#     df.to_csv('output.csv')

```

# Excel > to Dict > To API > to DF > To Excel

```python
#1) Read Excel - DONE
ex = pd.read_excel('For_test.xlsx').iloc[:,0].tolist()
print(type(ex))
print(ex)
```

```python
#2) Function - Get Answer - DONE
def query_excel(query):
    base_url = "https://trackapi.nutritionix.com//v2/search/instant"
    headers = {
        'x-app-id': '43fe6470',
        'x-app-key': 'bbb256572f595d38693c0c5dc462d5be'}
    body = {
        'query': query}
    r2 = requests.post(base_url, params=headers, data=body)
    r2_json = r2.json()
    dict1 = {}
    for i in r2_json:
        for l in r2_json[i]:
            dict1[i]=l
    return dict1
```

```python
query_excel('Cherry')
```

```python
#3) Iterate on Uploaded Excel values with API function - DONE
dict2 = {}
for i in ex:
    dict2[i]=query_excel(i)
#    print(i)
```

```python
#4) Convert Dictionary Output to Excel - DONE
dict_df = pd.DataFrame.from_dict({(i,j): dict2[i][j]
                                 for i in dict2.keys()
                                 for j in dict2[i].keys()},
                                orient='index')
dict_df
```
